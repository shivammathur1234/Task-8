```python
salary=[2533,3632,4562,4740,4856,5176,6666,7444]
exp=[2,3,4,5,6,7,8,9]
```

```python
plt.plot(salary,exp,marker="D")
plt.show()
```

```python
plt.plot(salary,exp,marker="^")
plt.show()
```

```python
plt.plot(salary,exp,marker="^",mfc='red')
plt.show()
```

```python
plt.plot(salary,exp,marker="^",mfc='red',mec='black')
plt.show()
```

```python
plt.plot(salary,exp,marker="^",mfc='red',mec='black',markersize='10')
plt.show()
```

```python
plt.plot(salary,exp,marker="^",mfc='red',mec='black',markersize='10',c='green')
plt.show()
```

```python
plt.plot(salary,exp,marker="^",mfc='red',mec='black',markersize='10',c='green',linewidth='10')
plt.show()
```

```python
plt.plot(salary,exp,marker="^",mfc='red',mec='black',markersize='10',c='green',linewidth='10')
plt.xticks([20000,30000,40000])
plt.yticks([2,6,8,12,15,20,21,])
plt.show()
```

```python
plt.plot(salary,exp,marker="^",mfc='red',mec='black',markersize='10',c='green',linewidth='10')
plt.xticks([20000,30000,40000])
plt.yticks([2,6,8,12,15,20,21,])
plt.xlabel('salary')
plt.show()
```

```python
plt.plot(salary,exp,marker="^",mfc='red',mec='black',markersize='10',c='green',linewidth='10')
plt.xticks([20000,30000,40000])
plt.yticks([2,6,8,12,15,20,21,])
plt.xlabel('salary')
plt.ylabel('Expenses')
plt.show()
```

```python
plt.plot(salary,exp,marker="^",mfc='red',mec='black',markersize='10',c='green',linewidth='10')
plt.title("salary vs Expenses",c='green')
plt.xticks([20000,30000,40000])
plt.yticks([2,6,8,12,15,20,21,])
plt.xlabel('salary')
plt.ylabel('Expenses')
plt.show()
```
